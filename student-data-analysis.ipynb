```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.width',500)
```

```python
# Source: https://www.kaggle.com/datasets/joebeachcapital/students-performance
df = pd.read_csv("dataset/DATA.csv")
df.head()
```

```python
df.rename(columns={
    'STUDENT ID': 'student_id',
    '1': 'age',
    '2': 'gender',
    '3': 'graduated_high_school_type',
    '4': 'scholarship_type',
    '5': 'additional_work',
    '6': 'regular_sports_activity',
    '7': 'is_partner',
    '8': 'total_salary',
    '9': 'transport',
    '10': 'accommodation',
    '11': 'mother_education',
    '12': 'father_education',
    '13': 'sibling_number',
    '14': 'parent_status',
    '15': 'mum_profession',
    '16': 'father_profession',
    '17': 'weekly_working_hours',
    '18': 'non-scientific_reading_frequency',
    '19': 'scientific_reading_frequency',
    '20': 'conference_participation',
    '21': 'project_success_impact',
    '22': 'lesson_continue',
    '23': 'midterm_exam_preparation_team',
    '24': 'midterm_exam_preparation_time',
    '25': 'course_note_taking',
    '26': 'lesson_listening',
    '27': 'discussion_lesson_impact',
    '28': 'flip_class',
    '29': 'past_gpa',
    '30': 'gpa',
    'COURSE ID': 'course_id',
    'GRADE': 'grade'
}, inplace=True)
```

### Number of people by age range View results

```python
age_group=["22-25", "18-21", "26-"]
age_intervals = df['age'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=age_group, y=age_intervals, palette='viridis', hue=age_intervals)
plt.title("Yas gruplarına göre öğrenci sayisi")
plt.xlabel("Yas aralıkları")
plt.ylabel("Kisi sayısı")
plt.show()
```

### Number of people by gender View results

```python
gender_gore_distribution=df["gender"].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=["Erkek","Kadın"], y=gender_gore_distribution, palette='viridis', hue=gender_gore_distribution)
plt.title("Cinsiyete göre dağılım")
plt.xlabel("Cinsiyet")
plt.ylabel("Kişi sayısı")

plt.show()
```

### Number of sexes by each age group View results

```python
for age_group in age_intervals.index:
    mourning_data = df[df['age'] == age_group]
    gender_distribution = mourning_data["gender"].value_counts()

    plt.figure(figsize=(12, 6))
    sns.barplot(x=gender_distribution, y=["Erkek", "Kadın"], palette="viridis", hue=gender_distribution)
    plt.title(f'{age_group.numerator} Yaş Grubu Cinsiyet Dağılımı')
    plt.xlabel("Cinsiyet")
    plt.ylabel("Kişi Sayısı")
    plt.show()
```

### How many students graduated from which type of high school

```python
high_school_group_name=["Devlet", "Özel", "Diğer"]
high_school_group=df['graduated_high_school_type'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=high_school_group_name, y=high_school_group, palette="viridis", hue=high_school_group)
plt.xlabel("Lise türü")
plt.ylabel("Kişi sayısı")
plt.show()
```

### Number of students according to scholarship amounts

```python
scholarship_kisi=df["scholarship_type"].value_counts()
scholarship_amount=["50", "75", "100", "25", "0"]

sns.barplot(x=scholarship_amount, y=scholarship_kisi, palette='viridis', hue=scholarship_kisi)
plt.xlabel("Burs miktarları")
plt.ylabel("Kişi sayısı")
plt.show()
```
